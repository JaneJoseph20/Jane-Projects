```python
# Importing packages
import pandas as pd
from IPython.display import display #Support for interactive data visualization
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore') 
```

```python
# Loading final dataset
finaldf = pd.read_csv('final version file.csv',encoding='ISO-8859-1')
finaldf.head(3)
```

```python
def myfunction():
    switch = True
    while switch:
        #Setting the menu options
        print("""
            \n'FRIENDS' TV show Analysis Menu
             Press 1 to see:  Top 3 episodes based on IMDB Ratings
             Press 2 to see:  Average IMDB Ratings of each season
             Press 3 to see:  Viewership fluctuations over the seasons
             Press 4 to see:  Top 15 directors based on the number of episodes they directed
             Press 5 to see:  Top 3 episodes based on Viewership
             Press 6 to quit
             """)      
       
        Choice = input("\nWhat's your choice?")
        
        try:
            Choice = int(Choice)
        except:
            print(f"\nYou entered an invalid option {Choice}. Please press 1,2,3,4,5 or 6")
            continue
        
        if Choice not in [1,2,3,4,5,6]:
            print(f"\nYou entered an invalid option {Choice}. Please press 1,2,3,4,5 or 6")
        else:
            if Choice == 1:
                print(f"You selected option {Choice}")
                print("\nTop 3 episodes based on IMDB Ratings")
                Choice1 = finaldf[['season', 'episode_num_in_season','title', 'imdb_rating']].sort_values(by='imdb_rating', ascending=False).head(3).style.hide_index()
                display(Choice1)
            if Choice == 2:
                print(f"You selected option {Choice}")
                plt.subplots(figsize=(10,6))
                plt.title('Average IMDB Rating of each season', fontweight ='bold', fontsize = 15)
                best_season = finaldf.groupby('season')['imdb_rating'].agg(['mean'])
                ax = sns.heatmap(best_season, annot=True, cmap="YlGnBu", cbar=True)
                ax.plot()
                plt.show()
                #ax = sns.heatmap(best_season, annot=True, cmap="YlGnBu", cbar=True)
                #display(ax)
            if Choice == 3:
                print(f"You selected option {Choice}")
                Season_viewership = finaldf[['us_viewers']].groupby(finaldf.season).mean().add_prefix('mean_')
                plt.rcParams["figure.figsize"] = (12,5)
                ax = Season_viewership.plot(title='Viewership fluctuations over the seasons').set(ylabel='Views in Millions', xlabel='Seasons')
                plt.show()
            if Choice == 4:
                print(f"You selected option {Choice}")
                plt.figure(figsize=(18,10)) 
                sns.set(font_scale=1.5)
                ax = sns.countplot(data=finaldf, y = 'directed_by', order = finaldf['directed_by'].value_counts().iloc[:15].index)
                plt.title('Top 15 directors based on the number of episodes they directed', fontsize=18)
                ax.plot()
                plt.show()
            if Choice == 5:
                print(f"You selected option {Choice}")
                print("Top 3 episodes based on Viewership")           
                Choice5 = finaldf[['season', 'episode_num_in_season','title', 'us_viewers']].sort_values(by='us_viewers', ascending=False).head(3).style.hide_index()
                display(Choice5)
            if Choice == 6:
                print(f"You selected option {Choice}. \nYou opted to quit.\nGoodbye.\nRun the function again to see the options")
                break
    switch == False      
```

```python
myfunction()
```
